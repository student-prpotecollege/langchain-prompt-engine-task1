```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

```python
# Task 1
# Converting hardcoded prompt into PromptTemplate

template1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners."
)

output1 = template1.format(topic="Artificial Intelligence")

print("Task 1 Output:")
print(output1)
```

```python
# Task 2
# Prompt with multiple inputs: topic, audience, tone

template2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

print("Task 2 Output 1:")
print(template2.format(topic="AI", audience="beginners", tone="friendly"))

print("\nTask 2 Output 2:")
print(template2.format(topic="Python", audience="kids", tone="fun"))

print("\nTask 2 Output 3:")
print(template2.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

```python
# Task 3
# Creating 3 different prompt templates

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step."
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}."
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story."
)

topic_name = "Machine Learning"

print("Task 3 - Teaching Prompt:")
print(teaching_template.format(topic=topic_name))

print("\nTask 3 - Interview Prompt:")
print(interview_template.format(topic=topic_name))

print("\nTask 3 - Storytelling Prompt:")
print(story_template.format(topic=topic_name))
```

```python
# Task 4
# ChatPromptTemplate using roles

role = "teacher"
topic = "Neural Networks"

if role == "teacher":
    system_message = "You are a helpful teacher who explains topics in a simple way."
elif role == "interviewer":
    system_message = "You are an interviewer who asks important questions."
elif role == "motivator":
    system_message = "You are a motivator who encourages students positively."
else:
    system_message = "You are a helpful assistant."

chat_template = ChatPromptTemplate.from_messages([
    ("system", system_message),
    ("user", "Tell me about {topic}.")
])

chat_output = chat_template.format_messages(topic=topic)

print("Task 4 Output:")
for message in chat_output:
    print(message)
```

```python
# Task 5
# Function to validate audience and tone

def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience given. Default value 'beginner' is assigned.")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone given. Default value 'formal' is assigned.")
        tone = "formal"

    return audience, tone


# Testing validation
print("Task 5 Output 1:")
print(validate_inputs("beginner", "fun"))

print("\nTask 5 Output 2:")
print(validate_inputs("kids", "friendly"))
```

```python
# Task 6
# Function to generate prompt based on style

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style."
        )

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask 3 interview questions about {topic} for {audience} in a {tone} tone."
        )

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style."
        )

    else:
        return "Invalid style. Please choose teaching, interview, or storytelling."

    return template.format(topic=topic, audience=audience, tone=tone)


print("Task 6 Output 1:")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))

print("\nTask 6 Output 2:")
print(generate_prompt("Python", "intermediate", "casual", "teaching"))

print("\nTask 6 Output 3:")
print(generate_prompt("Data Science", "expert", "formal", "interview"))
```

```python
# Task 7
# Using one template with 5 different inputs

reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

test_inputs = [
    {"topic": "Machine Learning", "audience": "beginner", "tone": "formal"},
    {"topic": "Python", "audience": "intermediate", "tone": "casual"},
    {"topic": "Neural Networks", "audience": "expert", "tone": "formal"},
    {"topic": "Data Analysis", "audience": "beginner", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "intermediate", "tone": "casual"}
]

print("Task 7 Outputs:")
for i, item in enumerate(test_inputs, start=1):
    print(f"\nOutput {i}:")
    print(reusable_template.format(
        topic=item["topic"],
        audience=item["audience"],
        tone=item["tone"]
    ))
```

```python
# Final pipeline:
# User Input -> Validation -> Prompt Template -> Dynamic Prompt Generation -> Output

topic = "Large Language Models"
audience = "beginner"
tone = "fun"
style = "teaching"

final_prompt = generate_prompt(topic, audience, tone, style)

print("Final Pipeline Output:")
print(final_prompt)
```

like topic, audience, tone, and style can be inserted dynamically.
